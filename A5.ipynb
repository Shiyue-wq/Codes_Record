```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize

q2 = "/Users/duilzhang/Downloads/NightLightData.csv"
df = pd.read_csv(q2)
df_sz = df[df['city']=='Shenzhen']
df_sv = df[df['city']!='Shenzhen']

df_sz['years'] = list(range(1993, 2014))
df_sz.rename(columns={'mean_nightlight':'nightlight_sz'}, inplace=True)
df_sv['years'] = list(range(1993, 2014))
df_sv.rename(columns={'mean_nightlight':'nightlight_sv'}, inplace=True)

df = pd.merge(df_sz, df_sv,on='years', how='outer' )

plt.figure(figsize=(10,6))
plt.plot(df['years'], df['nightlight_sz'], marker='o', linestyle='-', color='b',label='Shenzhen')
plt.plot(df['years'], df['nightlight_sv'], marker='o', linestyle='-', color='r',label='Silicon Valle')
plt.title("The average value of nighttime lights between Shenzhen and Silicon Valley from 1993-2013")
plt.grid(True)
plt.legend() 
plt.xlabel('Years')
plt.xticks(ticks=df['years'], labels=df['years'].astype(int), rotation=45) 
plt.savefig('nightlight.png')
plt.show()
plt.close()
```
