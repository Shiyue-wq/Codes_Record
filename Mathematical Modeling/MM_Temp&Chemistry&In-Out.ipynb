```python
import numpy as np
from matplotlib import pyplot as plt

def euler_rule(x0,y0,dt,f,max):
    x_val = [x0]
    y_val = [y0]
    while x0 < max:
        y0 += dt*f(x0,y0)
        x0 += dt
        x_val.append(x0)
        y_val.append(y0)
        
    return np.array(x_val),np.array(y_val)


def draw(x_val,y_val):
    plt.figure(figsize=(10, 6))
    plt.plot(x_val, y_val, label="Euler's Approximation", marker='o', linestyle='-', color='b')
    plt.savefig('Q1')
    plt.show()
    plt.close()
```

```python
#Question 1
x0 = 0
y0 = 100
dt = 0.5
max = 20

def f(x,y):
    k = 0.5
    return k*(20-y)

x_val,y_val = euler_rule(x0,y0,dt,f,max)
draw(x_val,y_val)
```

```python
#Question 2 
A0 = 50
B0 = 30
C0 = 0
x0 = 0
dt = 0.1

k1 = 0.5
k2 = 0.3 
k3 = 0.2


A_val = [A0]
B_val = [B0]
C_val = [C0]
x_val = [x0]

while x0 < 20:
    dA = (-k1*A0 + k3*C0)*dt
    dB = (-k2*B0 + k1*A0)*dt
    dC = (-k3*C0 + k2*B0)*dt
    x0 += dt
    A0 += dA
    B0 += dB
    C0 += dC
    A_val.append(A0)
    B_val.append(B0)
    C_val.append(C0)
    x_val.append(x0)
plt.figure(figsize=(10, 6))
plt.plot(x_val, A_val, label='[A] Concentration')
plt.plot(x_val, B_val, label='[B] Concentration')
plt.plot(x_val, C_val, label='[C] Concentration')
plt.xlabel('Time')
plt.ylabel('Concentration')
plt.title('Concentration of A, B, and C over Time using Euler\'s Method')
plt.legend()
plt.grid(True)
plt.savefig('Concentration')
plt.show()
plt.close()
```

```python
#Question 3
dt = 0.1  
time = np.arange(0, 100, dt)  
Q1 = [25] 
Q2 = [15]  

for t in time[:-1]:  
    Q1_val = 1.5 - Q1[-1]/10 + 3*Q2[-1]/40  
    Q2_val = 3 + Q1[-1]/10 - Q2[-1]/5       
    
    Q1_new = Q1[-1] + Q1_val * dt
    Q2_new = Q2[-1] + Q2_val * dt

    Q1.append(Q1_new)
    Q2.append(Q2_new)


time = time[:len(Q1)]


def draw3(time, a, b):
    plt.plot(time, a, label='Tank 1')
    plt.plot(time, b, label='Tank 2')
    plt.xlabel('Time (s)')
    plt.ylabel('Salt Amount (oz)')
    plt.legend()
    plt.title('The Amount of Salt in Each Tank Over Time')
    plt.grid(True)
    plt.savefig('Salt_in_tanks.png')  
    plt.show()  
    plt.close()  

draw3(time, Q1, Q2)
```
