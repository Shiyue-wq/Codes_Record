```python
import numpy as np
from sklearn.linear_model import LinearRegression

data = np.genfromtxt('/Users/duilzhang/Library/CloudStorage/OneDrive-CUHK-Shenzhen/DDA3020/HW1/linear_regression_train.txt', delimiter=' ', skip_header=0) 
X = data[:,:-1]
y = data[:, -1]
test = np.genfromtxt('/Users/duilzhang/Library/CloudStorage/OneDrive-CUHK-Shenzhen/DDA3020/HW1/linear_regression_test.txt', delimiter=' ', skip_header=0)
X_test = test

model = LinearRegression()
model.fit(X, y)
w0 = model.intercept_
coe = model.coef_
predictions = model.predict(X_test)

coefficients_dict = {f"w{i+1}": [w] for i, w in enumerate(coe)}
print('w0:',w0)
print(coefficients_dict)
print('y_hat:',predictions)
```

```python
ones_column = np.ones((X.shape[0], 1)) 
Xb = np.hstack([ones_column, X])
print('Xb:',Xb)
Xb_T = Xb.T
K = np.dot(Xb_T, Xb)
print('Xb_T*Xb:',K)
K_inv = np.linalg.inv(K)
print('K^(-1):',K_inv)
w = K_inv @ Xb_T @ y
w_dict = {f"w{i}": [w] for i, w in enumerate(w)}

one_col = np.ones((X_test.shape[0], 1)) 
Xtb = np.hstack([one_col, X_test])
y_predict = Xtb @ w
print(w_dict)
print('y_hat:',y_predict)
```
