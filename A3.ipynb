```python
import numpy as np
import matplotlib.pyplot as plt

#Excercise 1 
def bass_trout(alpha,beta,title):
    b0 = 5
    t0 = 2
    x0 = 0
    max = 2
    dx = 0.01
    
    b_val = [b0]
    t_val = [t0]
    x_val = [x0]
    while x0 < max:
        b0 += (b0*(10-b0-t0) - alpha*b0**2)*dx
        t0 += (t0*(15-b0-3*t0) - beta*t0**2)*dx
        x0 += dx
        b_val.append(b0)
        t_val.append(t0)
        x_val.append(x0)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x_val, b_val, label='Bass')
    plt.plot(x_val, t_val, label='Trout')
    plt.xlabel('Time')
    plt.ylabel("Bass&Trout's population")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.savefig(title)
    plt.show()
    plt.close()
    
    return
    
    
bass_trout(0,0,"Case1:alpha=beta=0")
bass_trout(1,2,"Case2:alpha=1, beta=2")
```

```python
#Excercise 2
def predator_prey(alpha,beta):
    x0 = 0.5
    y0 = 1
    t0 = 0
    max = 1
    dt = 0.01
    
    x_val = [x0]
    y_val = [y0]
    t_val = [t0]
    
    while t0 < max:
        x0 += dt*(x0-x0*y0-0.75*x0-alpha*x0**2)
        y0 += dt*(x0*y0-y0-0.75*y0-beta*y0**2)
        t0 += dt
        x_val.append(x0)
        y_val.append(y0)
        t_val.append(t0)
    
    return np.asarray(x_val), np.asarray(y_val), np.asarray(t_val)


#No competition within species:

#Competition within species
alpha = [0,0.5,1,0.3]
beta = [0,0.3,0.3,0.5]
x,y,t = [],[],[]

for i in range(len(alpha)):
    xi, yi, ti = predator_prey(alpha[i],beta[i])
    x.append(xi)
    y.append(yi)
    t.append(ti)

plt.figure(figsize=(10, 6))
for i in range(len(x)):
    plt.plot(t[i], x[i], label=f"Prey:alpha={alpha[i]},beta={beta[i]}")
    plt.plot(t[i], y[i], label=f"Predator:alpha={alpha[i]},beta={beta[i]}")
plt.legend()
plt.grid(True)
plt.savefig("Predator-Prey model with harvesting")
plt.show()
plt.close()
```
